```python
# Dataset: https://www.kaggle.com/datasets/mkechinov/ecommerce-behavior-data-from-multi-category-store/data
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./archive/2019-Oct.csv')
df
```

```python
df['event_type'].unique()
eventtype_count = df['event_type'].value_counts()
print(eventtype_count)
plt.bar(eventtype_count.index, eventtype_count.values)
plt.xlabel('Event_type')
plt.ylabel('Count')
plt.show()
```

```python
df['category_code'].unique()
category_count = df['category_code'].value_counts()
print(category_count)
plt.bar(category_count.index, category_count.values)
plt.xlabel('category_code')
plt.ylabel('Count')
plt.show()
```

```python
df['event_time'] = pd.to_datetime(df['event_time'])

views = df[df['event_type'] == 'view'].sort_values(by='event_time')
purchases = df[df['event_type'] == 'purchase']

merged_df = pd.merge(purchases, views, on=['user_id', 'product_id'], suffixes=('_purchase', '_view'))

merged_df = merged_df[merged_df['event_time_view'] < merged_df['event_time_purchase']]

views_before_purchase = merged_df.groupby(['user_id', 'product_id', 'event_time_purchase']).size().reset_index(name='views_before_purchase')

views_before_purchase.sort_values(by='event_time_purchase', inplace=True)
```

```python
count = 0
for user_id, group in views_before_purchase.groupby('user_id'):
    if count < 20:
        if(len(group['event_time_purchase']) < 5):
            continue
        print(len(group['event_time_purchase']))
        plt.figure(figsize=(10, 6))
        plt.plot(group['event_time_purchase'], group['views_before_purchase'], marker='o', linestyle='-')
        plt.title(f'Views Before Purchase Over Time for User {user_id}')
        plt.xlabel('Purchase Time')
        plt.ylabel('Number of Views Before Purchase')
        plt.grid(True)
        plt.show()
        count+=1
    else:
        break
```

```python

df['event_time'] = pd.to_datetime(df['event_time'])

views = df[df['event_type'] == 'view'].sort_values(by='event_time')
purchases = df[df['event_type'] == 'purchase']

merged_df = pd.merge(purchases[['user_id', 'product_id', 'price', 'event_time']], views[['user_id', 'product_id', 'event_time']], on=['user_id', 'product_id'], suffixes=('_purchase', '_view'))

merged_df = merged_df[merged_df['event_time_view'] < merged_df['event_time_purchase']]

views_before_purchase = merged_df.groupby(['user_id', 'product_id', 'event_time_purchase', 'price']).size().reset_index(name='views_before_purchase')


grouped = merged_df.groupby(['user_id', 'product_id', 'event_time_purchase', 'price']).size().reset_index(name='views_before_purchase')

def user_data(group):
    correlation = group['views_before_purchase'].corr(group['price'])
    avg_price = group['price'].mean()
    avg_views = group['views_before_purchase'].mean()
    return pd.Series({'correlation': correlation, 'avg_price': avg_price, 'avg_views': avg_views})

user_correlation_data = grouped.groupby('user_id').apply(user_data).reset_index()

user_correlation_data = user_correlation_data.dropna(subset=['correlation'])

print(user_correlation_data)


plt.figure(figsize=(12, 6))
plt.plot(user_correlation_data['user_id'], user_correlation_data['correlation'], marker='o', linestyle='-', color='b')
plt.title('Correlation Between Views Before Purchase and Purchase Price per User')
plt.xlabel('User ID')
plt.ylabel('Correlation Coefficient')
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to make room for label rotation
plt.show()
```
